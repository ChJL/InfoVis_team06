```python
w1 = 1
w2 = 2
w3 = 3
w4 = 4

data=pd.read_csv('dataset.csv')

def Weighted(w1,w2,w3,w4):
    #Normalized
    tw = w1+w2+w3+w4
    W1 = w1/tw
    W2 = w2/tw
    W3 = w3/tw
    W4 = w4/tw
    
    Scores = data.EPI*W1+data.COL*W2+data.International*W3+data.Quality*W4
    Subset = list(zip(data.Country,Scores))
    df = pd.DataFrame(data = Subset, columns=['Country', 'Scores'])
    Sorted = df.sort_values(['Scores'], ascending=False)
    Sorted['Rank'] = range(1,32)
    return Sorted

print(Weighted(w1,w2,w3,w4))
```
